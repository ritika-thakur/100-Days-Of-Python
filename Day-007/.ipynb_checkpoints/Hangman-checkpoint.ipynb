# Hangman

```python
import random
from hangman_art import stages, logo
from hangman_art import word_list
```

```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
```

```python
word_list = ["aardvark", "baboon", "camel", "koala", "porcupine", "sloth"]
display = []
chosen_word = random.choice(word_list)

print(f"Solution: {chosen_word}")
for i in chosen_word:
    display += "_"
print(display)

lives=6 

while (lives>0):
    
    guess = input("Guess a letter: ").lower()
    for position in range(len(chosen_word)):
        if guess == chosen_word[position]:
            display[position] = guess
        else:
            continue    
    
    if guess not in display:
        lives = lives -1
    if "_" not in display:
        print("You win! You saved a life today.")
    if lives == 0:
        print("You lose.") 
        
    print("Lives Left: ",lives)
    print(display)
    print(stages[lives])
```
