```python
from data import data
from art import logo, vs 
import random
```

```python
def celeb_info():
    celeb = random.randint(0, len(data))
    name = data[celeb]['name']
    description = data[celeb]['description']
    country =  data[celeb]['country']
    follower_count = data[celeb]['follower_count']
    return name, description, country, follower_count
    

correct_answer = True
score = 0
while correct_answer:
    print(logo)
    A_name, A_description, A_country, A_follower_count = celeb_info()
    print(f"Comparision A: {A_name}, a {A_description}, from {A_country}.")
    print(vs)
    B_name, B_description, B_country, B_follower_count = celeb_info()
    print(f"Against B: {B_name}, a {B_description}, from {B_country}.")
    guess = input("Who has more followers? Type 'A' or 'B' : ")
    if guess == 'A' and A_follower_count > B_follower_count:
        score += 1
        print(f"You are right! Current Score: {score}")
        
    elif guess == 'B' and A_follower_count < B_follower_count:
        score += 1
        print(f"You are right! Current Score: {score}")
        
    else: 
        print(f"Sorry that's wrong! Final Score: {score}")
        correct_answer = False
        
```
