############### Our Blackjack House Rules #####################
#.The deck is unlimited in size. 
#.There are no jokers. 
#.The Jack/Queen/King all count as 10.
#.The the Ace can count as 11 or 1.
#.Use the following list as the deck of cards:
#.cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
#.The cards in the list have equal probability of being drawn.
#.Cards are not removed from the deck as they are drawn.


```python
import random
from art import logo
print(logo)
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

def distribute_cards():
    players_deck = [ ]
    comp_deck = []
    for _ in range(0,2):
        players_card = random.choice(cards)
        players_deck.append(players_card)
        comp_card = random.choice(cards)
        comp_deck.append(comp_card)
    return players_deck, comp_deck


def draw_cards(p_deck, c_deck):
    p_score = sum(p_deck)
    while p_score < 21:
        p_deck.append(random.choice(cards))
        another_card = input("Type 'y' to get another card, type 'n' to pass: ")
        p_score = sum(p_deck)
        print(f"Your final hand: {p_deck}, final score: {p_score}")
        if another_card == 'n':
            break
    c_score = sum(c_deck)        
    if p_score > 21:
        return "You went over. You lose."
    elif p_score > c_score:
        return "Congratulations! You win"
    else:
        return "You lose! Computer win"
        

def result(p_deck, c_deck):
    p_score = sum(p_deck)
    c_score = sum(c_deck)
    
    if p_score > 21:
        result = "You went over. You lose."
    elif p_score < 21 and c_score < 21:
        if p_score > c_score:
            count = 0
            while p_score > c_score:
                c_deck.append(random.choice(cards))
                c_score = sum(c_deck)
            if c_score > 21:
                result = (f"Congratulations you win!")
            elif c_score == p_score:
                result = "Draw!"
            else:
                result = (f"You lose. Computer won.")
        elif p_score == c_score: 
            result = ("Draw!")
        elif p_score < c_score:
            another_card = input("Type 'y' to get another card, type 'n' to pass: ")
            if another_card == 'y':
                result = draw_cards(p_deck, c_deck)
            else:
                result = "You lose. computer won"
        else:
            result = ("You lose. Computer won.")
    else:
        result = "Congratulation! You won the game."
    
    print(f"Your final hand: {p_deck}, final score: {p_score}")
    print(f"Computer's final hand: {c_deck}, final score: {c_score}")
    print(result)
    
next_game = True
while next_game:
    play_again = input("Do you want to play a game of Blackjack? Type 'y' or 'n' ")
    if play_again == 'n':
        next_game = False
        print("GoodBye")
        continue
    players_current_cards, comp_current_cards = distribute_cards()
    players_score = sum(players_current_cards)
    print(f"Your cards: {players_current_cards}, current score: {players_score}")
    print(f"Computer's first card: {comp_current_cards[0]}")
    if players_score == 21:
        print("BLACKJACK! You won.")
        continue
    another_card = input("Type 'y' to get another card, type 'n' to pass: ")
    if another_card == 'y':
        draw_cards(players_current_cards, comp_current_cards)
        result(players_current_cards, comp_current_cards)
    else:
        result(players_current_cards, comp_current_cards)
```
